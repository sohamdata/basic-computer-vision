# reading images

```python
img = cv2.imread(r"images\jumbo.jpg")
```

```python
img
```

```python
img.shape
#1600x1600x3 (three channels)
```

## extracting the rgb values of a pixel

```python
(B,G,R) = img[100,100]
print(f'R={R}, G={G}, B={B}')
```

```python
# opencv reads image in BGR format, coverting it into RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
plt.imshow(img2)
```

# extract the region of intrest

```python
roi = img2[200:900, 600:1200]
plt.imshow(roi)
```

```python
#resizing an image
resize = cv2.resize(img2, (800,800))
resize.shape
```

# blurring the image

```python
img_blur = cv2.blur(img,(20,30))
cv2.imshow("blur", img_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# edge detection in an image

```python
edges = cv2.Canny(img2, 200,200)
cv2.imshow("original", img2)
cv2.waitKey(0)
cv2.imshow("edge", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## image filtering

```python
# gaussian filter- linear, takes support of supporting pixels of the object
# median filter- non-linear, preserves edges

noise = cv2.imread(r"images\noisy_image.png")
cv2.imshow("original", noise)
cv2.waitKey(0)
gaussian = cv2.GaussianBlur(noise,(5,5),0)
cv2.imshow("gaussian", gaussian)
cv2.waitKey(0)
median = cv2.medianBlur(noise,5)
cv2.imshow("median", median)
cv2.waitKey(0)
```

```python
desc = cv2.putText(img, "nice kitty", (0,150), cv2.FONT_HERSHEY_SIMPLEX, 5,(0,0,225),5)
plt.imshow(desc)
```
